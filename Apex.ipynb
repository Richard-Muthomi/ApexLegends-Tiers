## Apex Tiers

```python
import constraint
damage=0
problem=constraint.Problem()
problem.addVariable("Wraith","damage=350")
problem.addVariable("Bangalore","damage=500")
problem.addVariable("Lifeline","damage=300")
problem.addVariable("Gibby","damage=600")

def my_constraint(wraith,bangalore,lifeline,gibby):
    if damage<350:
        print("B tier")
    elif damage>=400:
        print("S tier")

problem.addConstraint(my_constraint,["Wraith","Bangalore","Lifeline","Gibby"])
solutions=problem.getSolutions()

for s in solutions:
    print(s)
    
```
